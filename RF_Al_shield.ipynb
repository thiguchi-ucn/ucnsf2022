```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
df0c =  pd.read_csv('data_RF/AluminiumRFShielding/NoShielding/Y=0mmPlot.txt', comment='%', sep='\s+', names=['length', '|B|']) 
df0off =  pd.read_csv('data_RF/AluminiumRFShielding/NoShielding/Y=42.5mmPlot.txt', comment='%', sep='\s+', names=['length', '|B|']) 

df0off['B_G'] = df0off['|B|']*1e4
df0off['z_mm'] = (df0off['length']-28)*10
df0c['B_G'] = df0c['|B|']*1e4
df0c['z_mm'] = (df0c['length']-28)*10

fig0, ax0 = plt.subplots(2,2, figsize=(10,8), sharex=True, )

ax0[0,0].plot(df0c['z_mm'], df0c['B_G'], lw=1, label='without shield', c='k')
ax0[1,0].plot(df0c['z_mm'], df0c['B_G'], lw=1, label='without shield', c='k')
ax0[0,1].plot(df0off['z_mm'], df0off['B_G'], lw=1, label='without shield', c='k')
ax0[1,1].plot(df0off['z_mm'], df0off['B_G'], lw=1, label='without shield', c='k')
for i in range(2):
    ax0[1,i].set_ylim(0, 0.25)
    ax0[0,i].set_ylim(0, 3.5)
    ax0[0,i].grid('both', ls='-.')
    ax0[1,i].grid('both', ls='-.')
    ax0[0,i].axvline(-64, lw=1, ls='-.', color='k')
    ax0[1,i].axvline(-64, lw=1, ls='-.', color='k')
    ax0[0,i].axvline(-150, lw=1, ls='-.', color='k')
    ax0[1,i].axvline(-150, lw=1, ls='-.', color='k')

# ls_list = [2*i for i in range(1, 20)]
ls_list = [4*i-2 for i in range(1, 11)]
for ls in  ls_list:
    
    
    df_i = pd.read_csv('data_RF/AluminiumRFShielding/3mmThick%dmmLength/Y=0mmPlot.txt' %ls, comment='%', sep='\s+', names=['length', '|B|'])
    df_i['B_G'] = df_i['|B|']*1e4
    df_i['z_mm'] = (df_i['length']-28)*10
        
    df_ioff = pd.read_csv('data_RF/AluminiumRFShielding/3mmThick%dmmLength/Y=42.5mmPlot.txt' %ls, comment='%', sep='\s+', names=['length', '|B|'])
    df_ioff['B_G'] = df_ioff['|B|']*1e4
    df_ioff['z_mm'] = (df_ioff['length']-28)*10
    
    ax0[0,0].plot(df_i['z_mm'], df_i['B_G'], lw=1, label='shield length %d mm' %ls)
    ax0[1,0].plot(df_i['z_mm'], df_i['B_G'], lw=1, label='shield length %d mm' %ls)
    ax0[0,1].plot(df_ioff['z_mm'], df_ioff['B_G'], lw=1, label='shield length %d mm' %ls)
    ax0[1,1].plot(df_ioff['z_mm'], df_ioff['B_G'], lw=1, label='shield length %d mm' %ls)
for i in range(2):
    for j in range(2):
        ax0[i,j].legend(fontsize=8, loc='upper right')
        ax0[i,j].set_ylabel('Re(Bz) (G)')
        ax0[i,j].axvline(-64, lw=1, ls='-', color='gray')
        ax0[i,j].axvline(64, lw=1, ls='-', color='gray')
        ax0[i,j].axvline(-150, lw=1, ls='-', color='gray')
        ax0[i,j].axvline(150, lw=1, ls='-', color='gray')
        ax0[i,j].axvline(-64+15, lw=1, ls='-.', color='gray')
        ax0[i,j].axvline(-64-15, lw=1, ls='-.', color='gray')
        ax0[i,j].axvline(64+15, lw=1, ls='-.', color='gray')
        ax0[i,j].axvline(64-15, lw=1, ls='-.', color='gray')
        ax0[1,j].set_xlabel('z (mm)')
fig0.savefig('Al_result1')
```
